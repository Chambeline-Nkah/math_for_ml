## Task 1: Write a dynamic function to find the derivative of any function f(x)

```python
#importing the SymPy python library which is used for symbolic mathematics.
import sympy as sp
```

```python
def dy_dx(equation, var = "x"):
    x = sp.symbols(var) #Defining the symblic variable 'x'
    f = sp.sympify(equation) #Converting equation which is a string to a SymPy expression
    der = sp.diff(f, x) #Differentiating f with respect to x
    return der #Returning the resulting derivative
```

## Task 2: Test the derivative function written with a quadratic equation of your choice

```python
equation = "x**2 - x + 1"
derivative_of_eqn = dy_dx(equation)
print(derivative_of_eqn)
```
